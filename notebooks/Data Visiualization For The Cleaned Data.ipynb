```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('D:/downloads 2 imp/2ND TERM 3RD YEAR/Intelligent Programming/MAIN PROJECT/data/heart_Cleaned data.csv')

# Set aesthetic style
sns.set_theme(style="whitegrid")

#Statistical Summary(describtive analysis)
summary = df.describe()
print("\nstatistical Summary: ",summary)
```

```python
#Correlation Heatmap to represent the relation between features 
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
```

```python
# Define the features to plot (excluding the target itself)
features = df.columns.drop('target')

# Create a grid of plots
fig, axes = plt.subplots(nrows=len(features) // 3 + 1, ncols=3, figsize=(20, 25))
axes = axes.flatten()

for i, col in enumerate(features):
    sns.histplot(
        data=df, 
        x=col, 
        hue='target', 
        multiple="stack", # Use "dodge" for side-by-side or "stack" for vertical overlap
        palette={0: "skyblue", 1: "salmon"}, 
        ax=axes[i], 
        kde=True # Adds a trend line
    )
    axes[i].set_title(f'Distribution of {col} by Target')

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#Boxplots after removing the outliers using iqr in the cleaning phase
plt.figure(figsize=(15, 8))
sns.boxplot(data=df.drop(columns=['target']), palette='Set2')
plt.title('Boxplots of Features (Check for Outliers)')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Calculate the absolute correlation of each feature with the target
# We drop 'target' itself to only look at predictors
significance = df.corr()['target'].abs().sort_values(ascending=False).drop('target')

#Visualize the significance ranking
plt.figure(figsize=(10, 6))
sns.barplot(x=significance.values, y=significance.index, palette='magma')

plt.title('Feature Significance (Absolute Correlation with Target)')
plt.xlabel('Correlation Magnitude (0 to 1)')
plt.ylabel('Features')
plt.tight_layout()
```
